```python
import re
import csv
from json import loads

filegraph = "graph.csv"
jsonfile = "TwitterCrawl.json"

content = {}
targets = {}
emitters = {}
adjacency = {}

i = 0

with open(jsonfile, 'r') as f:
    data = f.readlines()
    for line in data:
        try:
            line_object = loads(line)
            content[i] = line_object["text"]
            emitters[i] = line_object["user"]["screen_name"]
            targets[i] = re.findall("@([a-zA-Z0-9_]{1,15})", str(content[i]))
            i = i + 1
        except:
            pass
        
targets
```

```python
length = max(targets.values(), key=len)
print(len(length))
```

```python
upper_limit = len(length)
i = 0        
for value in emitters.values():
    adjacency[value] = targets[i]   
    i = i + 1

i = 0    
with open(filegraph, 'w') as f:
    writer = csv.writer(f)
    for key, value in adjacency.items():
        try:
            if (value):
                for i in range(0,(upper_limit+1)): 
                    if (value[i]):
                        writer.writerow([key, value[i]])
            else:
                writer.writerow([key])
        except:
            pass
```
